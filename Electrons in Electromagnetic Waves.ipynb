## 用数值方法求解电子在平面电磁波中的运动(相对论性)

相关公式如下 
* $\vec{E}=E_{0}(z-c t) \sin \left(\omega t-\frac{\omega}{c} z\right) \hat{x}$ 
* $\vec{B}=\frac{1}{c} E_{0}(z-c t) \sin \left(\omega t-\frac{\omega}{c} z\right) \hat{y}$
* $E_{0}(z-c t)$  是平面波的 (纵向) 包络
* 例如  $E_{0} \propto e^{-\left(\frac{z-c t-z_{0}}{c \tau}\right)^{2}}, \tau$  是激光脉宽
* $\frac{d}{d t} \vec{p}=-e\left(\vec{E}+\frac{\vec{p}}{m_{e} \gamma} \times \vec{B}\right)$ 
* $\gamma=\sqrt{1+\left(\frac{\vec{p}}{m_{e} c}\right)^{2}}$

#### 任意给定电子初始条件(位置,动量),观察$\gamma$随时间的变化,检查激光包络完全通过后$\gamma$与初始条件的差别

#### 典型的激光参数:波长800纳米,$E_0$峰值$4 × 10^{12} V/m$。可改变E0峰值查看对电子轨迹的影响。

利用四阶Runge-Kutta方法求解微分方程 \
设方程为
\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
\frac{\mathrm{d} x_1}{\mathrm{d}t}  = f_1(t,x_1,x_2,...,x_m) \\
\frac{\mathrm{d} x_2}{\mathrm{d}t}  = f_2(t,x_1,x_2,...,x_m) \\
... \\
\frac{\mathrm{d} x_m}{\mathrm{d}t}  = f_m(t,x_1,x_2,...,x_m) \\
\end{array}\right. \\
\end{aligned}
\end{equation}
其中初值为
\begin{equation}
\begin{aligned}
x_{1}(0) = x_{1}^{0}, x_{2}(0) = x_{2}^{0}, \ldots, x_{m}(0) = x_{m}^{0}
\end{aligned}
\end{equation}

四阶Runge-Kutta方法可以表示为
\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
x^{k+1}_1  = x^{k}_1 +\frac{1}{6}(f_{11}+2f_{12}+2f_{13}+f_{14}) \\
x^{k+1}_2  = x^{k}_2 +\frac{1}{6}(f_{21}+2f_{22}+2f_{23}+f_{23}) \\
... \\
x^{k+1}_m  = x^{k}_m +\frac{1}{6}(f_{m1}+2f_{m2}+2f_{m3}+f_{m4}) \\
\end{array}\right. \\
k = 0,1,2,3...
\end{aligned}
\end{equation}

其中
\begin{equation}
\begin{aligned}
f_{i1} &= f_i \left( t_k, x_1^k, x_2^k, \dots, x_m^k \right) \\
f_{i2} &= f_i \left( t_k + \frac{h}{2}, x_1^k + \frac{h}{2} f_{11}, x_2^k + \frac{h}{2} f_{21}, \dots, x_m^k + \frac{h}{2} f_{m1} \right) \\
f_{i3} &= f_i \left( t_k + \frac{h}{2}, x_1^k + \frac{h}{2} f_{12}, x_2^k + \frac{h}{2} f_{22}, \dots, x_m^k + \frac{h}{2} f_{m2} \right) \\
f_{i4} &= f_i \left( t_k + h, x_1^k + h f_{13}, x_2^k + h f_{23}, \dots, x_m^k + h f_{m3} \right)\\
i &= 1, 2, 3... m 
\end{aligned}
\end{equation}

写出电子运动方程的分量形式：

\begin{cases}
\frac{\mathrm{d} p_x}{\mathrm{d}t}  = -eE_x +\frac{e}{m_e \gamma} p_z B_y \\
\frac{\mathrm{d} p_y}{\mathrm{d}t}  = 0 \\
\frac{\mathrm{d} p_z}{\mathrm{d}t}  = -\frac{e}{m_e \gamma} p_x  B_y \\
\end{cases}
考虑到粒子的坐标变化，加入方程：
\begin{cases}
\frac{\mathrm{d} x}{\mathrm{d}t}  = \frac{p_x}{m_e \gamma} \\
\frac{\mathrm{d} y}{\mathrm{d}t}  = \frac{p_y}{m_e \gamma} \\
\frac{\mathrm{d} z}{\mathrm{d}t}  = \frac{p_z}{m_e \gamma} \\
\end{cases}

其中
$$
\gamma=\sqrt{1+\left(\frac{p_x^2 + p_y^2 + p_z^2}{m_{e} c}\right)^{2}}
$$

将电子运动方程的分量形式代入Runge-Kutta方法中，得到有关动量和位置的六个方程：
\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
f_{11}  = -eE_t(t^{(k)},z^{(k)}) + \frac{e}{m_e \sqrt{1+\frac{p_x^{2^{(k)}} + p_y^{2^{(k)}} + p_z^{2^{(k)}}}{(m_{e} c)^{2}}}} p_z^{(k)} B_y(t^{(k),z^{(k)}}) \\
f_{21}  = 0 \\
f_{31}  =  -\frac{e}{m_e \sqrt{1+\frac{p_x^{2^{(k)}} + p_y^{2^{(k)}} + p_z^{2^{(k)}}}{(m_{e} c)^{2}}}} p_x^{(k)} B_y(t^{(k),z^{(k)})}\\
f_{41}  =  \frac{1}{m_e \sqrt{1+\frac{p_x^{2^{(k)}} + p_y^{2^{(k)}} + p_z^{2^{(k)}}}{(m_{e} c)^{2}}}} p_x^{(k)} \\
f_{51}  =  \frac{1}{m_e \sqrt{1+\frac{p_x^{2^{(k)}} + p_y^{2^{(k)}} + p_z^{2^{(k)}}}{(m_{e} c)^{2}}}} p_y^{(k)} \\
f_{61}  =  \frac{1}{m_e \sqrt{1+\frac{p_x^{2^{(k)}} + p_y^{2^{(k)}} + p_z^{2^{(k)}}}{(m_{e} c)^{2}}}} p_z^{(k)} \\
\end{array}\right. \\
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
f_{12}  = -eE_t(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{61}) + \frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_z^{(k)}+ \frac{h}{2}f_{31}) B_y(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{61}) \\
f_{22}  = 0 \\
f_{32}  =  -  \frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_x^{(k)}+ \frac{h}{2}f_{11}) B_y(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{61})\\
f_{42}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}  (p_x^{(k)}+ \frac{h}{2}f_{11}) \\
f_{52}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}(p_y^{(k)}+ \frac{h}{2}f_{21}) \\
f_{62}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_z^{(k)}+ \frac{h}{2}f_{31}) \\
\end{array}\right. \\
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
f_{13}  = -eE_t(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{62}) +\frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_z^{(k)}+ \frac{h}{2}f_{32}) B_y(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{62}) \\
f_{23}  = 0 \\
f_{33}  =  \frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_x^{(k)}+ \frac{h}{2}f_{12}) B_y(t^{(k)}+\frac{h}{2},z^{(k)}+\frac{h}{2}f_{62})\\
f_{43}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}  (p_x^{(k)}+ \frac{h}{2}f_{12}) \\
f_{53}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}  (p_y^{(k)}+ \frac{h}{2}f_{22}) \\
f_{63}  = \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}  (p_z^{(k)}+ \frac{h}{2}f_{32}) \\
\end{array}\right. \\
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
\left\{\begin{array}{l}
f_{14}  = -eE_t(t^{(k)}+h,z^{(k)}+hf_{62}) + \frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_z^{(k)}+ hf_{32}) B_y(t^{(k)}+h,z^{(k)}+hf_{62}) \\
f_{24}  = 0 \\
f_{34}  =  -  \frac{e}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_x^{(k)}+ hf_{12}) B_y(t^{(k)}+h,z^{(k)}+hf_{62})\\
f_{44}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}(p_x^{(k)}+ hf_{13}) \\
f_{54}  =  \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}} (p_y^{(k)}+ hf_{22}) \\
f_{64}  =   \frac{1}{m_e \sqrt{1+\frac{(p_x^{(k)}+ hf_{12})^2 + (p_x^{(k)}+ hf_{22})^2 + (p_z^{(k)}+ hf_{32})^2}{(m_{e} c)^{2}}}}(p_z^{(k)}+ hf_{32}) \\
\end{array}\right. \\
\end{aligned}
\end{equation}

```python
import numpy as np
import scipy.constants as constant
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

考虑：
* $\vec{E}=E_0 \sin \left(\omega t-\frac{\omega}{c} z\right) \hat{x}$ 
* $\vec{B}=\frac{1}{c} E_0 \sin \left(\omega t-\frac{\omega}{c} z\right) \hat{y}$

将$E_0,\omega$设为参数

```python
def electron_motion(px0, py0, pz0, x0, y0, z0,E0,lambda_laser,dt,t):
    '''
    This function calculates the motion of an electron in a laser field
    using the Runge-Kutta method.

    Parameters:
    px0, py0, pz0, x0, y0, z0 is the initial momentum and position of the electron.
    E0 is the amplitude of the electric field.
    lambda_laser is the wavelength of the laser field.
    dt is the time step, t is the total time.

    Returns:
    x, y, z, px, py, pz and gamma is the position and momentum of the electron at each time step.
    
    '''
    c = constant.c
    m_e = constant.m_e
    e = constant.e
    omega = 2 * np.pi * c / (lambda_laser*10**(-9)) # calculate the angular frequency of the laser field
    
    info = {'px':[], 'py':[], 'pz':[], 'x':[], 'y':[], 'z':[], 'gamma':[], 't':[]}  # store the information of the electron
    n = int(t/dt) # number of time steps
    x_k = x0; y_k = y0; z_k = z0; px_k = px0; py_k = py0; pz_k = pz0  # initial position and momentum
    gamma = np.sqrt( 1 + (px_k**2 + py_k**2 + pz_k**2) / (m_e * c)**2) # calculate the Lorentz factor
    info['px'].append(px_k); info['py'].append(py_k); info['pz'].append(pz_k); 
    info['x'].append(x_k); info['y'].append(y_k); info['z'].append(z_k);
    info['gamma'].append(gamma); info['t'].append(0)  # store the information of the electron at t = 0


    
    for i in range(n): 

        t_k   = info['t'][-1]          # get the time at the time i*dt
        x_k   = info['x'][-1] 
        y_k   = info['y'][-1] 
        z_k   = info['z'][-1]  # get the position of the electron at the time i*dt
        px_k  = info['px'][-1] 
        py_k  = info['py'][-1] 
        pz_k  = info['pz'][-1]  # get the momentum of the electron at the time i*dt
        gamma = info['gamma'][-1]  # get the Lorentz factor of the electron at the time i*dt
        h = dt /2 # calculate the half time step

        
        # calculate the Runge-Kutta coefficients f_i1
        E = E0 * np.sin(omega * t_k - omega / c * z_k)     # calculate the electric field 
        B = E0 / c * np.sin(omega * t_k - omega / c * z_k) # calculate the magnetic field
        f11 = -e * E +  (e / (m_e * gamma)) * pz_k *B
        f21 = 0
        f31 = (e / (m_e * gamma)) * px_k * B
        f41 = (1 / (m_e * gamma)) * px_k
        f51 = (1 / (m_e * gamma)) * py_k
        f61 = (1 / (m_e * gamma)) * pz_k


        # calculate the Runge-Kutta coefficients f_i2
        # calculate the process variables 
        px_k = info['px'][-1] + h * f11
        py_k = info['py'][-1] + h * f21
        pz_k = info['pz'][-1] + h * f31
        x_k = info['x'][-1] + h * f41
        y_k = info['y'][-1] + h * f51
        z_k = info['z'][-1] + h * f61
        t_k = info['t'][-1] + h
        E = E0 * np.sin(omega * t_k - omega / c * z_k)      # calculate the electric field 
        B = E0 / c * np.sin(omega * t_k - omega / c * z_k)  # calculate the magnetic field  
        # 计算 f 值
        f12 = -e * E +  (e / (m_e * gamma)) * pz_k *B
        f22 = 0
        f32 = -(e / (m_e * gamma)) * px_k * B
        f42 = (1 / (m_e * gamma)) * px_k
        f52 = (1 / (m_e * gamma)) * py_k
        f62 = (1 / (m_e * gamma)) * pz_k


        # calculate the Runge-Kutta coefficients f_i3
        px_k = info['px'][-1] + h * f12
        py_k = info['py'][-1] + h * f22
        pz_k = info['pz'][-1] + h * f32
        x_k  = info['x'][-1]  + h * f42
        y_k  = info['y'][-1]  + h * f52
        z_k  = info['z'][-1]  + h * f62
        t_k  = info['t'][-1]  + h
        E = E0 * np.sin(omega * t_k - omega / c * z_k)     # calculate the electric field 
        B = E0 / c * np.sin(omega * t_k - omega / c * z_k) # calculate the magnetic field   
        # 计算 f 值
        f13= -e * E +  (e / (m_e * gamma)) * pz_k *B
        f23= 0
        f33= -(e / (m_e * gamma)) * px_k * B
        f43= (1 / (m_e * gamma)) * px_k
        f53= (1 / (m_e * gamma)) * py_k
        f63= (1 / (m_e * gamma)) * pz_k


        # calculate the Runge-Kutta coefficients f_i4
        px_k = info['px'][-1] + dt * f13
        py_k = info['py'][-1] + dt * f23
        pz_k = info['pz'][-1] + dt * f33
        x_k  = info['x'][-1]  + dt * f43
        y_k  = info['y'][-1]  + dt * f53
        z_k  = info['z'][-1]  + dt * f63
        t_k  = info['t'][-1]  + dt
        E = E0 * np.sin(omega * t_k - omega / c * z_k)     # calculate the electric field 
        B = E0 / c * np.sin(omega * t_k - omega / c * z_k) # calculate the magnetic field    
        # 计算 f 值
        f13= -e * E +  (e / (m_e * gamma)) * pz_k *B
        f23= 0
        f33= -(e / (m_e * gamma)) * px_k * B
        f43= (1 / (m_e * gamma)) * px_k
        f53= (1 / (m_e * gamma)) * py_k
        f63= (1 / (m_e * gamma)) * pz_k

        # calculate the position and momentum of the electron at the time i*dt
        px_k = info['px'][-1] + dt / 6 * (f11 + 2 * f12 + 2 * f13 + f13)
        py_k = info['py'][-1] + dt / 6 * (f21 + 2 * f22 + 2 * f23 + f23)
        pz_k = info['pz'][-1] + dt / 6 * (f31 + 2 * f32 + 2 * f33 + f33)
        x_k  = info['x'][-1]  + dt / 6 * (f41 + 2 * f42 + 2 * f43 + f43)
        y_k  = info['y'][-1]  + dt / 6 * (f51 + 2 * f52 + 2 * f53 + f53)
        z_k  = info['z'][-1]  + dt / 6 * (f61 + 2 * f62 + 2 * f63 + f63)
        # calculate the Lorentz factor of the electron at the time i*dt
        gamma = np.sqrt( 1 + (px_k**2 + py_k**2 + pz_k**2) / (m_e * c)**2)
        


        # store the information of the electron
        info['t'].append(t_k)
        info['px'].append(px_k)
        info['py'].append(py_k)
        info['pz'].append(pz_k)
        info['x'].append(x_k)
        info['y'].append(y_k)
        info['z'].append(z_k)
        info['gamma'].append(gamma)

    return info
```

```python
em = electron_motion(0., 0., 0., 0., 0., 0.,4 * 10**12, 800  ,1*10**-6,0.1)
```

```python

t = em['t']  

# 定义粒子坐标的函数
x = em['x']  # x坐标关于时间的函数
y = em['y']  # y坐标关于时间的函数
z = em['z']  # z坐标关于时间的函数

# 创建三维图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制轨迹
ax.plot(x, y, z, label='Particle trajectory', color='blue')

# 添加标签和标题
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Trajectory of the Particle')
ax.legend()

# 显示图形
plt.show()
```
